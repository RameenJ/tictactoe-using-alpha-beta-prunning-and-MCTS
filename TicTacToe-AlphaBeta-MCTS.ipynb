```python
from random import choice
from math import inf
def abminimax(board, depth, alpha, beta, player):
    row = -1
    col = -1
    if depth == 0 or gameWon(board):
        return [row, col, getScore(board)]

    else:
        for cell in blanks(board): #for blanks positions: cell[x,y]
            setMove(board, cell[0], cell[1], player) # player = 1, X // Here we are simultaing moves
            score = abminimax(board, depth - 1, alpha, beta, -player) # recursive call, so it will keep simulating player moves
            if player == 1: # in our case, ye opponent hai
                # X is always the max player
                #updating alpha for maximizer
                if score[2] > alpha:
                    alpha = score[2]
                    row = cell[0]
                    col = cell[1]

            else:
              #updating beta for minimizer
                if score[2] < beta:
                    beta = score[2]
                    row = cell[0]
                    col = cell[1]
            # Revert the cell's value to 0 to explore other possible moves
            setMove(board, cell[0], cell[1], 0)

            #pruning
            if alpha >= beta:
                break

        if player == 1:
            return [row, col, alpha]

        else:
            return [row, col, beta]

def o_comp(board):

    #making random move if it is the first move of the game
    if len(blanks(board)) == 9:
        x = choice([0, 1, 2]) #choice function imported from random module
        y = choice([0, 1, 2])
        setMove(board, x, y, -1)
        Gameboard(board)

    # using abminmax to make smart move
    else:
        result = abminimax(board, len(blanks(board)), -inf, inf, -1)
        setMove(board, result[0], result[1], -1) # abmin max returns a list [row,col,alpha/beta]
        Gameboard(board)

def x_comp(board):
    if len(blanks(board)) == 9:
        x = choice([0, 1, 2])
        y = choice([0, 1, 2])
        setMove(board, x, y, 1)
        Gameboard(board)

    else:
        result = abminimax(board, len(blanks(board)), -inf, inf, 1)
        setMove(board, result[0], result[1], 1)
        Gameboard(board)

def makeMove(board, player, mode):
    if mode == 1:
        if player == 1:
            playerMove(board)

        else:
            o_comp(board)
    else:
        if player == 1:
            o_comp(board)
        else:
            x_comp(board)

#player vs computer game
def pvc():
    while True:
        try:
            order = int(input('Enter to play 1st or 2nd: '))
            if not (order == 1 or order == 2):
                print('Please pick 1 or 2')
            else:
                break
        except(KeyError, ValueError):
            print('Enter a number')

    Clearboard(board)
    if order == 2:
        currentPlayer = -1
    else:
        currentPlayer = 1

    while not (boardFull(board) or gameWon(board)):
        makeMove(board, currentPlayer, 1)
        currentPlayer *= -1

    printResult(board)


# computer vs computer game
def cvc():

    Clearboard(board)
    o_comp(board)
    while not (boardFull(board) or gameWon(board)):
        x_comp(board)
        if not (boardFull(board)):
          o_comp(board)



    printResult(board)


```

```python
import random #ye waala
import math
board = [[0, 0, 0],
         [0, 0, 0],
         [0, 0, 0]]

def Gameboard(board):
    chars = {1: 'X', -1: 'O', 0: ' '}
    for x in board:
        for y in x:
            ch = chars[y]
            print(f'| {ch} |', end='')
        print('\n' + '---------------')
    print('===============')

def Clearboard(board):
    for x, row in enumerate(board):
        for y, col in enumerate(row):
            board[x][y] = 0

def winningPlayer(board, player):
    conditions = [[board[0][0], board[0][1], board[0][2]],
                     [board[1][0], board[1][1], board[1][2]],
                     [board[2][0], board[2][1], board[2][2]],
                     [board[0][0], board[1][0], board[2][0]],
                     [board[0][1], board[1][1], board[2][1]],
                     [board[0][2], board[1][2], board[2][2]],
                     [board[0][0], board[1][1], board[2][2]],
                     [board[0][2], board[1][1], board[2][0]]]

    if [player, player, player] in conditions:
        return True

    return False

def gameWon(board):
    return winningPlayer(board, 1) or winningPlayer(board, -1)

def printResult(board):
    if winningPlayer(board, 1):
        print('X has won! ' + '\n')

    elif winningPlayer(board, -1):
        print('O\'s have won! ' + '\n')

    else:
        print('Draw' + '\n')

def blanks(board):
    blank = []
    for x, row in enumerate(board):
        for y, col in enumerate(row):
            if board[x][y] == 0:
                blank.append([x, y])

    return blank

def boardFull(board):
    if len(blanks(board)) == 0:
        return True
    return False

def setMove(board, x, y, player):
    board[x][y] = player

def playerMove(board):
    e = True
    moves = {1: [0, 0], 2: [0, 1], 3: [0, 2],
             4: [1, 0], 5: [1, 1], 6: [1, 2],
             7: [2, 0], 8: [2, 1], 9: [2, 2]}
    while e:
        try:
            move = int(input('Enter a number between 1-9: '))
            if move < 1 or move > 9:
                print('Invalid Move! Try again!')
            elif not (moves[move] in blanks(board)):
                print('Invalid Move! Try again!')
            else:
                setMove(board, moves[move][0], moves[move][1], 1)
                Gameboard(board)
                e = False
        except(KeyError, ValueError):
            print('Enter a number!')

def getScore(board):
    if winningPlayer(board, 1):
        return 10

    elif winningPlayer(board, -1):
        return -10

    else:
        return 0
class TreeNode:
    def __init__(self, board, player):
        self.board = [row[:] for row in board]
        self.player = player
        self.children = []
        self.parent = None
        self.visits = 0
        self.score = 0

def ucb(node):
    if node.visits == 0:
        return float('inf')
    return (node.score / node.visits) + 1.41 * math.sqrt(math.log(node.parent.visits) / node.visits)

def select(node):
    if not node.children:
        return node
    return select(max(node.children, key=ucb))

def expand(node):
    legal_moves = blanks(node.board)
    for move in legal_moves:
        child_board = [row[:] for row in node.board]
        setMove(child_board, move[0], move[1], node.player)
        child = TreeNode(child_board, -node.player)
        node.children.append(child)
        child.parent = node
    return random.choice(node.children)

def evaluate(board, player):
    # Heuristic evaluation function
    score = 0
    opponent = -player

    # Check rows, columns, and diagonals for potential wins
    for i in range(3):
        if [player, player, player] in [board[i], [board[0][i], board[1][i], board[2][i]]]:
            score += 100
        if [opponent, opponent, opponent] in [board[i], [board[0][i], board[1][i], board[2][i]]]:
            score -= 100

    if [player, player, player] in [board[0][0], board[1][1], board[2][2]]:
        score += 100
    if [opponent, opponent, opponent] in [board[0][0], board[1][1], board[2][2]]:
        score -= 100

    if [player, player, player] in [board[0][2], board[1][1], board[2][0]]:
        score += 100
    if [opponent, opponent, opponent] in [board[0][2], board[1][1], board[2][0]]:
        score -= 100

    return score

def simulate(node):
    board = [row[:] for row in node.board]
    player = node.player
    while not gameWon(board) and not boardFull(board):
        legal_moves = blanks(board)
        if random.random() < 0.9:
            # Select a move based on the evaluation function with some exploration
            move_scores = []
            for move in legal_moves:
                temp_board = [row[:] for row in board]
                setMove(temp_board, move[0], move[1], player)
                move_scores.append((move, evaluate(temp_board, player)))
            move_scores.sort(key=lambda x: -x[1])
            move = move_scores[0][0]
        else:
            move = random.choice(legal_moves)
        setMove(board, move[0], move[1], player)
        player = -player
    return getScore(board)

def backpropagate(node, result):
    node.visits += 1
    node.score += result
    if node.parent is not None:
        backpropagate(node.parent, -result)

def uct_search(board, player, iterations):
    root = TreeNode(board, player)
    for _ in range(iterations):
        node = select(root)
        if gameWon(node.board) or boardFull(node.board):
            result = getScore(node.board)
        else:
            if not node.children:
                expand(node)
            child = random.choice(node.children)
            result = simulate(child)
        backpropagate(node, result)
    best_child = max(root.children, key=lambda child: child.visits)
    return best_child.board

if __name__ == "__main__":
    board = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
    Gameboard(board)

    while not gameWon(board) and not boardFull(board):
        playerMove(board)
        if gameWon(board) or boardFull(board):
            break

        board = uct_search(board, -1, 50000)  # Increase the number of iterations for more intelligent play
        Gameboard(board)

    printResult(board)
```

```python
import random
import math

class Node:
    def __init__(self, state, parent=None):
        self.state = state
        self.parent = parent
        self.children = []
        self.wins = {1: 0, 2: 0}  # Track wins for each player
        self.visits = 0

    def is_fully_expanded(self):
        return len(self.children) == len(self.state.get_legal_moves())

    def select_child(self):
        for child in self.children:
            if child.visits == 0:
                return child
        total_visits = sum(child.visits for child in self.children)
        log_total = math.log(total_visits or 1)
        best_score = float("-inf")
        best_child = None
        for child in self.children:
            exploit = child.wins[self.state.current_player] / (child.visits or 1)
            explore = math.sqrt(log_total / (child.visits or 1))
            score = exploit + 1.41 * explore
            if score > best_score:
                best_score = score
                best_child = child
        return best_child

    def expand(self):
        legal_moves = self.state.get_legal_moves()
        for move in legal_moves:
            new_state = self.state.copy()
            new_state.make_move(move)
            new_node = Node(new_state, self)
            self.children.append(new_node)
        return self.select_child()

    def simulate(self):
        current_state = self.state.copy()
        while not current_state.is_terminal():
            random_move = random.choice(current_state.get_legal_moves())
            current_state.make_move(random_move)
        return current_state.get_winner()

    def backpropagate(self, result):
        self.visits += 1
        if result == 0:  # Draw
            self.wins[1] += 0.5
            self.wins[2] += 0.5
        else:
            self.wins[result] += 1
        if self.parent:
            self.parent.backpropagate(result)


class GameState:
    def __init__(self, board=None, current_player=1):
        if board is None:
            self.board = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
        else:
            self.board = board
        self.current_player = current_player

    def copy(self):
        new_state = GameState()
        new_state.board = [row[:] for row in self.board]
        new_state.current_player = self.current_player
        return new_state

    def get_legal_moves(self):
        moves = []
        for i in range(3):
            for j in range(3):
                if self.board[i][j] == 0:
                    moves.append((i, j))
        return moves

    def make_move(self, move):
        i, j = move
        self.board[i][j] = self.current_player
        self.current_player = 3 - self.current_player

    def is_terminal(self):
        if self.get_winner() is not None or len(self.get_legal_moves()) == 0:
            return True
        return False

    def get_winner(self):
        lines = (
            self.board[0],
            self.board[1],
            self.board[2],
            [self.board[i][0] for i in range(3)],
            [self.board[i][1] for i in range(3)],
            [self.board[i][2] for i in range(3)],
            [self.board[i][i] for i in range(3)],
            [self.board[i][2 - i] for i in range(3)],
        )
        for line in lines:
            if all(cell == 1 for cell in line):
                return 1
            elif all(cell == 2 for cell in line):
                return 2
        if all(self.board[i][j] != 0 for i in range(3) for j in range(3)):
            return 0  # Draw
        return None

    def print_board(self):
      for i in range(3):
          for j in range(3):
              print("|", end=" ")
              if self.board[i][j] == 0:
                  print("  ", end=" ")
              elif self.board[i][j] == 1:
                  print("X ", end=" ")
              elif self.board[i][j] == 2:
                  print("O ", end=" ")
          print("|")
          if i < 2:
              print("---------------")
      print("===============")


def mcts_search(state, iterations):
    root = Node(state)
    for _ in range(iterations):
        node = root
        while not node.is_fully_expanded() and not node.state.is_terminal():
            node = node.expand()
        if node.children:
            best_score = float("-inf")
            best_child = None
            exploration_param = 1.4
            for child in node.children:
                if child.visits == 0:
                    exploit = 0
                    explore = math.sqrt(2 * math.log(node.visits))
                else:
                    exploit = child.wins[node.state.current_player] / child.visits
                    explore = math.sqrt(2 * math.log(node.visits) / child.visits)
                score = exploit + exploration_param * explore
                if score > best_score:
                    best_score = score
                    best_child = child
            node = best_child
        simulation_result = node.simulate()
        node.backpropagate(simulation_result)
    best_child = max(root.children, key=lambda c: c.visits)
    return best_child.state

def main():
    game_state = GameState()
    while not game_state.is_terminal():
        if game_state.current_player == 1:
            game_state.print_board()
            row, col = map(int, input("Enter your move (row, col): ").split())
            while (row, col) not in game_state.get_legal_moves():
                print("Invalid move. Try again.")
                row, col = map(int, input("Enter your move (row, col): ").split())
            game_state.make_move((row, col))
        else:
            game_state = mcts_search(game_state, 5000)
    game_state.print_board()
    winner = game_state.get_winner()
    if winner == 1:
        print("You win!")
    elif winner == 2:
        print("AI wins!")
    else:
        print("It's a draw!")

if __name__ == "__main__":
    main()
```
